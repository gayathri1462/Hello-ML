```python
games=games[games["users_rated"]> 0]
```

```python
plt.hist(games["average_rating"],color='orange')
plt.xlabel('Average Rating')
plt.ylabel('Number of games')
plt.savefig('plot1.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(30,20))
corr = games.corr()
sns.heatmap(corr, annot=True, cmap=plt.cm.Reds)
plt.savefig('plot2.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.regplot(data=games, x="users_rated", y="average_rating",color="orange")
plt.savefig('plot3.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
cor_target = abs(corr["average_rating"])
relevant_features = cor_target[cor_target>0.1]
relevant_features
```

```python
X = games[['id', 'yearpublished','minage', 'users_rated', 'bayes_average_rating', 'total_owners',
       'total_traders', 'total_wanters', 'total_wishers', 'total_comments','total_weights', 'average_weight']]
Y = games["average_rating"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(X,Y,test_size =0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100, min_samples_leaf=10, random_state=1)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred= rf.predict(x_test)
y_pred
```

```python
mean_squared_error(y_pred, y_test)
```
